```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objs as go
```

```python
df1=pd.read_csv("data/러시아인원손실.csv")
러시아장비손실=pd.read_csv("data/러시아장비손실.csv")
```

```python
df1.head()
```

```python
## 러시아 군인 인원손실
fig=px.line(df1,x='date',y='personnel',markers=True,template='plotly',color_discrete_sequence=['crimson'],title='<b>러시아 군인 인원손실')
fig.update_traces(marker=dict(size=4))
fig.update_layout(font_color='crimson',title_x=0.5)
```

```python
#  러시아 항공무기 손실
date = 러시아장비손실['date']
aircraft = 러시아장비손실['aircraft']
helicopter = 러시아장비손실['helicopter']
anti = 러시아장비손실['anti-aircraft warfare']
drone = 러시아장비손실['drone']

# Create plot
air = go.Figure()

# Add traces
air.add_trace(go.Scatter(x=date, y=aircraft,
                    mode='lines+markers',
                    name='Aircraft'))
air.add_trace(go.Scatter(x=date, y=helicopter,
                    mode='lines+markers',
                    name='Helicopter'))
air.add_trace(go.Scatter(x=date, y=anti,
                    mode='lines+markers',
                    name='Anti-aircraft warfare'))
air.add_trace(go.Scatter(x=date, y=drone,
                    mode='lines+markers',
                    name='Drone'))
air.update_layout(legend_orientation="h",
                  legend=dict(x=0, y=1, traceorder="normal"),
                  title="항공무기",
                  xaxis_title="Date",
                  yaxis_title="Weapons ",
                  margin=dict(l=0, r=0, t=30, b=0))
air.show()
```

```python
# 러시아 총 장비 손실
dict_russia_losses_equipment_sum = 러시아장비손실.sum(axis=0)

layout = dict(title =  '러시아 장비 손실')


fig = go.Figure(data=[go.Pie(labels=dict_russia_losses_equipment_sum.keys(), 
                             values=dict_russia_losses_equipment_sum.values, 
                             textinfo='label',
                             insidetextorientation='radial'
)], layout=layout)

fig.show()
```
